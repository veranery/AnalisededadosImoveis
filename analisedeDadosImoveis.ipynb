**Criando colunas numéricas com os valores gastos mensais e anuais**


```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep=(';'))
dados

```

```python
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
dados.head()
```

```python
dados['Valor_por_ano'] = dados['Valor_por_mes'] * 12 + dados['IPTU']
dados.head()
```

```python
dados['Valor_por_mes'].describe()
```


Ao analisar os dados fornecidos, é possível identificar indícios de uma considerável variação nos valores pagos por mês. O desvio padrão, que está em torno de 57.125,54, sinaliza uma dispersão significativa em relação à média. Essa medida estatística reflete a magnitude das diferenças individuais em relação ao valor médio, indicando uma considerável amplitude nos pagamentos mensais.

A mediana, que é uma medida de tendência central robusta, mostra que metade dos valores está concentrada em torno de 3.790. Isso sugere uma distribuição na qual uma parte substancial dos dados está agrupada nessa faixa. No entanto, a média, que é mais sensível a valores extremos, está em 9.943,364, indicando que a presença de valores mais altos influencia essa média para cima.

Essa disparidade entre a mediana e a média fortalece a interpretação de que há uma grande variação nos valores pagos por mês. Enquanto a mediana representa um ponto central mais resistente a valores atípicos, a média é impactada por esses valores mais elevados, resultando em uma média consideravelmente maior.

Em resumo, a análise conjunta da mediana e da média sugere uma distribuição assimétrica com uma cauda mais longa em direção aos valores mais altos, corroborando a ideia de uma considerável variação nos pagamentos mensais.
